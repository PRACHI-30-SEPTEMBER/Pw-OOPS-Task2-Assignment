## Q1. Explain why we have to use the Exception class while creating a Custom Exception.

**Answer:**  
In Python, all built‑in exceptions are derived from the base class **Exception**. When we create a custom exception, we inherit from the Exception class so that our custom error behaves like standard Python errors.

Reasons:
- It allows the custom error to be caught using try–except blocks.
- It keeps the error hierarchy consistent with Python's built‑in exceptions.
- It provides built‑in features such as error messages and traceback.


```python
# Example of Custom Exception
class MyCustomError(Exception):
    pass

try:
    raise MyCustomError('This is a custom exception')
except MyCustomError as e:
    print('Caught Exception:', e)
```

## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

**Answer:**  
ArithmeticError is the base class for errors related to arithmetic operations.

Common errors under ArithmeticError:
- ZeroDivisionError
- OverflowError
- FloatingPointError

**Example 1: ZeroDivisionError** – occurs when a number is divided by zero.


```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print('Cannot divide by zero')
```

**Example 2: OverflowError** – occurs when a calculation exceeds the maximum limit.


```python
import math

try:
    print(math.exp(1000))
except OverflowError:
    print('Number too large')
```

## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

**Answer:**  
LookupError is the base class for errors that occur when a lookup operation fails. It mainly includes **IndexError** and **KeyError**.

- **IndexError:** occurs when accessing an invalid index in a list.
- **KeyError:** occurs when accessing a non‑existing key in a dictionary.


```python
# IndexError Example
try:
    data = [1,2,3]
    print(data[5])
except IndexError:
    print('Index does not exist')
```

```python
# KeyError Example
try:
    student = {'name':'Rahul'}
    print(student['age'])
except KeyError:
    print('Key not found in dictionary')
```

## Q5. Explain ImportError. What is ModuleNotFoundError?

**Answer:**

- **ImportError:** occurs when Python fails to import a module or a specific object from a module.
- **ModuleNotFoundError:** a subclass of ImportError that occurs when the module itself cannot be found.


```python
try:
    import non_existing_module
except ModuleNotFoundError:
    print('Module not found')
```

## Q6. List down some best practices for exception handling in Python.

**Answer:**

Best practices include:

- Use specific exception types instead of a generic except block.
- Avoid using bare except statements.
- Keep try blocks small and focused.
- Use finally to release resources like files or database connections.
- Log exceptions for debugging.
- Create custom exceptions when needed for better readability.
